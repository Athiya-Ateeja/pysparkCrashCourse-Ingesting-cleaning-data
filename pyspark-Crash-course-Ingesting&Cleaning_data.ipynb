```
df = spark.read.csv("dbfs:/FileStore/shared_uploads/athiyastudies@gmail.com/original.csv",header=True)
```

```
df.show()
```

```
from pyspark.sql.functions import *
df = df.withColumn("clean_city",when(df.City.isNull(),'Unknown').otherwise(df.City))
```

```
display(df)
```

```
df = df.filter(df.JobTitle.isNotNull())
```

```
df.show()
```

```
df = df.withColumn("clean_salary",df.Salary.substr(2,100).cast('float'))
```

```
df.show()
```

```
mean = df.groupBy().avg("clean_salary").take(1)[0][0]
```

```
print(mean)
```

```
from pyspark.sql.functions import lit
```

```
df = df.withColumn("New_sal",
                   when(df.clean_salary.isNull(),lit(mean))\
                        .otherwise(df.clean_salary))
```

```
df.show()
```

```
import numpy as np
```

```
Longitude = df.select("Longitude")
```

```
Longitude.show()
```

```
Longitude = Longitude.withColumn("Longitude2",Longitude.Longitude.cast('float')).select('Longitude2')
```

```
Longitude.show()
```

```
meadian = np.median(Longitude.collect())
```

```
print(meadian)
```

```
df = df.withColumn("lat",when(df.Longitude.isNull(),lit(meadian))\
                    .otherwise(df.Longitude))
```

```
df.show()
```
